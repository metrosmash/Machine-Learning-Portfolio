## Transforming The Data To A DataFrame
- so i will turn the log data into a dataframe and remove the useless values
- i will also turn it into a csv file for other uses 

```python
#open the file 
f = open('/kaggle/input/wordpress-ddos-log-dataset/Wordpress DDOS attack Logs.txt', "r") 

#initialize an empty list
mylist = [] 

#iterate through each line in the file
for line in f: 
  
    #strip the newline character 
    line = line.strip() 
  
    #append the line to the list 
    mylist.append(line) 
  
#close the file 
f.close() 
```

- This converts the log.txt file into a list 

```python
import re
list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
for line in mylist:
    word = str(line)
    parts = re.split(r'[\s\[\]\"]+', word)
    list1.append(parts[0])
    list2.append(parts[3])
    list3.append(parts[8:10])
    list4.append(parts[11])
    list5.append(parts[12])
```

- This line then converts the list into different lists representing the column of our data

```python
dict = {'IP adress':list1,
        'DateandTime':list2,
        'Port':list3,
        'Wordpress Version':list4,
        'Website': list5}
df = pd.DataFrame(dict)
display(df.head())
```

- Here  the data is turned into a pandas dataframe (more like a table)

```python
DateTimeSplit = df["DateandTime"].str.split(":", n = 1, expand = True)
DateTimeSplit
```

- split the date and time column into date column and time column

```python
df["DT"]= DateTimeSplit[0]
df["Time"]= DateTimeSplit[1]
display(df)
```

```python
df = df.drop('DateandTime',axis =1)
df['Time'] = pd.to_datetime(df['Time'], dayfirst=True).dt.time
display(df.head())
```

- The above code completes cleaning the data
- it makes date and time different columns and deletes the original DateandTime column
- Now the code as been fully converted to a usable dataframe 

```python
df.to_csv('wordpress_ddos_log_transform.csv')
```
